```python
from groq import Groq
import os

client = Groq(api_key=os.getenv("GROQ_API_KEY"))

BASE_MODEL = "llama-3.3-70b-versatile"
```

```python
MODEL_CONFIG = {
    "technical": {
        "system_prompt": (
            "You are a senior software engineer. "
            "Be precise, technical, and solution-oriented. "
            "Provide code snippets when relevant. "
            "Focus on debugging and implementation details."
        ),
        "temperature": 0.7
    },
    "billing": {
        "system_prompt": (
            "You are a customer support billing specialist. "
            "Be empathetic, polite, and policy-driven. "
            "Explain refund policies clearly and professionally."
        ),
        "temperature": 0.7
    },
    "general": {
        "system_prompt": (
            "You are a friendly general customer support assistant. "
            "Respond conversationally and helpfully."
        ),
        "temperature": 0.7
    }
}
```

```python
def route_prompt(user_input: str) -> str:
    routing_prompt = f"""
Classify the following user query into one of these categories:
[technical, billing, general]

Return ONLY the category name.

User Query:
{user_input}
"""

    response = client.chat.completions.create(
        model=BASE_MODEL,
        temperature=0,  # deterministic routing
        messages=[
            {"role": "system", "content": "You are a strict classifier."},
            {"role": "user", "content": routing_prompt}
        ]
    )

    category = response.choices[0].message.content.strip().lower()

    if category not in MODEL_CONFIG:
        return "general"

    return category
```

```python
def process_request(user_input: str) -> str:

    # Step 1: Route
    category = route_prompt(user_input)
    print("ðŸ”Ž Routed to:", category)

    # Step 2: Load expert config
    expert_config = MODEL_CONFIG[category]

    # Step 3: Call expert model
    response = client.chat.completions.create(
        model=BASE_MODEL,
        temperature=expert_config["temperature"],
        messages=[
            {"role": "system", "content": expert_config["system_prompt"]},
            {"role": "user", "content": user_input}
        ]
    )

    return response.choices[0].message.content
```

```python
query = "My python script is throwing an IndexError on line 5."
result = process_request(query)

print("\nFinal Response:\n")
print(result)
```
